```python
from bs4 import BeautifulSoup
import re

# Load your HTML file
with open("html\IPC-1-400.html", "r", encoding="utf-8") as file:
    html_content = file.read()

# Parse the HTML using BeautifulSoup
soup = BeautifulSoup(html_content, "html.parser")
```
