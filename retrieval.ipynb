```python
import pandas as pd
from collections import Counter

def create_subsection_entity_frequency_df(df):
    rows = []
    # Iterate over each record in the original DataFrame
    for idx, row in df.iterrows():
        # Get the raw ner_subsec_texts; assumed to be a list of tuples
        ner_subsec = row['ner_subsec_texts']
        if isinstance(ner_subsec, list):
            # Each tuple should be like: (subsec_id, nested_list)
            for subsec in ner_subsec:
                if isinstance(subsec, tuple) and len(subsec) >= 2:
                    subsec_id = subsec[0]
                    nested_list = subsec[1]
                    entities = []
                    # nested_list is assumed to be a list of lists containing tag dictionaries
                    if isinstance(nested_list, list):
                        for inner in nested_list:
                            if isinstance(inner, list):
                                for tag in inner:
                                    # Extract the entity if available
                                    if isinstance(tag, dict) and 'entity' in tag:
                                        entities.append(tag['entity'])
                    # Count frequency of each entity
                    freq = Counter(entities)
                    # Build a row with the parent record identifier, subsection id, and frequency counts
                    row_data = {'number': row['number'], 'subsec_id': subsec_id}
                    row_data.update(freq)
                    rows.append(row_data)
    # Create a new DataFrame from the collected rows
    freq_df = pd.DataFrame(rows)
    # Replace NaN values with 0 (in case some subsections don't have certain entities)
    freq_df = freq_df.fillna(0)
    return freq_df

# Example usage:
entity_freq_df = create_subsection_entity_frequency_df(df)
print(entity_freq_df.head())
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

def match_subsection_by_freq(input_freq, freq_df, top_n=1):
    """
    Given an input frequency dictionary (e.g., {'B-ORGANIZACAO': 3, 'I-ORGANIZACAO': 1, ...})
    and a frequency DataFrame (with columns 'number', 'subsec_id', and entity tags),
    this function computes cosine similarity between the input vector and each subsection vector,
    returning the top_n matching subsections along with their similarity scores.
    """
    # Identify tag columns (exclude identifier columns)
    tag_columns = [col for col in freq_df.columns if col not in ['number', 'subsec_id']]
    
    # Build the input vector in the same column order, using 0 if a tag is missing in input_freq
    input_vector = np.array([input_freq.get(tag, 0) for tag in tag_columns]).reshape(1, -1)
    
    # Extract the frequency matrix for each subsection
    matrix = freq_df[tag_columns].values
    
    # Compute cosine similarity between the input vector and each subsection vector
    similarities = cosine_similarity(input_vector, matrix).flatten()
    
    # Identify indices for the top_n most similar subsections
    top_indices = np.argsort(similarities)[-top_n:][::-1]
    
    # Create a result DataFrame with the matching rows and add the similarity scores
    result = freq_df.iloc[top_indices].copy()
    result['similarity'] = similarities[top_indices]
    
    return result

# Example usage:
# Assume `entity_freq_df` is the DataFrame created using create_subsection_entity_frequency_df(df)
# And an example input frequency dictionary:
input_frequency = {'B-ORGANIZACAO': 2, 'I-ORGANIZACAO': 1, 'B-LOCAL': 3}
top_matches = match_subsection_by_freq(input_frequency, entity_freq_df, top_n=5)
print(top_matches)
```
